dei o seguinte comando no terminal:

```bash
mlflow models serve -m 'runs:/54342d61b4aa4691b09136f60306ef41/model' -p 5002 --env-manager=local
```

Isso foi feito para iniciar um servidor REST API local na porta 5002, que hospeda o modelo treinado salvo no MLflow.

**Servir o modelo como uma API**: Permite que outros sistemas, aplicações ou scripts enviem dados para o endpoint `/invocations` e recebam as predições do modelo em tempo real.

Quando usar?

- Quando você quer disponibilizar seu modelo para consumo externo (por exemplo, por aplicações web, scripts, sistemas de produção, etc).
- Para testar o modelo via API sem precisar escrever código Python para carregar e rodar o modelo.

````python
import requests
````

````python
url = "http://127.0.0.1:5002/invocations"
````

````python
data = {
    "dataframe_split": {
        "columns": ["tamanho", "ano", "garagem"],
        "data": [[159.0, 2003, 2]],
    }
}
````

````python
header = {"Content-Type": "application/json"}
````

````python
response = requests.post(url, json=data, headers=header)
response
````

````python
print(response.text)
````
