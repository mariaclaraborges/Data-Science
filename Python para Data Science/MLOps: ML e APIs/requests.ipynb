```python
import requests
```

```python
url = "http://127.0.0.1:5000/cotacao/"
```

```python
dados = {
    "tamanho": 120,
    "ano": 2001,
    "garagem": 1
}
```

```python
auth = requests.auth.HTTPBasicAuth('admin', 'admin')
```

```python
response = requests.post(url, json=dados, auth=auth)
```
